```python
!pip install --force-reinstall numpy==1.23.5
import urllib.request

urllib.request.urlretrieve(
    r"https://raw.githubusercontent.com/thomasnield/machine-learning-demo-data/master/llm/legend_of_sleepy_hollow.txt", 
    "legend_of_sleepy_hollow.txt"
)

filename = 'legend_of_sleepy_hollow.txt' 
file = open(filename, encoding="utf-8")
text = file.read()
file.close()
text
```

```python
# display the text 
text
```

```python
import re 

text = re.sub(r"^(.|\n)+START OF THE PROJECT GUTENBERG EBOOK THE LEGEND OF SLEEPY HOLLOW \*{3}", '', text)
text = re.sub(r"\*{3} END OF THE PROJECT GUTENBERG EBOOK THE LEGEND OF SLEEPY HOLLOW (.|\n)+", '', text)
text = text.strip()

text
```

```python
#Manual Tokenization
text.split()
```

```python
# We can again use regular expressions to match whitespace or more elaborate patterns. 
# In this case, hyphenated words are split into separate tokens.
```

```python
import re 

words = re.split(r'\W+', text)

words
```

```python
# Now let's say we want to remove punctuation. We can get a convenient set 
# of punctuation characters from Python's standard library.
import re 
import string 

print(string.punctuation)
```

```python
# We can then construct a character set using a regular expression by 
# using these punctuation characters, and remove said punctuation characters.
regex_punct = re.compile(f'[{re.escape(string.punctuation)}]')
stripped = [regex_punct.sub('', w) for w in words]
stripped
```

```python
# We probably should concern ourselves with making the casing consistent, as in uppercase 
# or lowercase and making sure one convention is stuck to.
lowercased = [w.lower() for w in stripped]
lowercased
```

```python
#Downloading collection 'all'
# import nltk
# nltk.download('all')
import nltk
nltk.download()
```

```python
#Breaking Up Words
# We can split up words in NLTK using the word_tokenize() function. It will split on white space 
# and punctuation including commas, periods, and contractions like what's -> what 's
from nltk.tokenize import word_tokenize
words = word_tokenize(text)
words
```

```python
# You will see here that the tokens above have punctuation marks as separate tokens. 
# We can filter those out if we like using the is_alpha() function.
no_puncts = [w for w in words if w.isalpha()]
no_puncts
```

```python
#Breaking Up Sentences
# Another way we can process this text is to break it up into sentences rather than words. 
# We can bring in the sent_tokenize() function from NLTK to achieve this. We can then grab the 25th sentence in the story.
from nltk import sent_tokenize

sentences = sent_tokenize(text)
print(sentences[25])
```

```python
# Removing stop words, which are words that bear little meaning like
# the and is. You look at stopwords available for English in NLTK. 
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
# We can take these stop words, package them into a set, and remove them from our text. 
# Note because the stop words are in lower case, we should compare each word in lower case as well.
no_stop_words = [w for w in no_puncts if not w.lower() in stop_words]
no_stop_words
```

Stemming

```python
# There might be times you want to reduce each word to its root or base. 
# The words fighter and fighting stem from fight. This can help reduce the vocabulary and find broader tones or sentiments in the document. 
# The most popular stemming algorithm is the Porter Stemming algorithm which NLTK has available.
from nltk.stem.porter import PorterStemmer

porter = PorterStemmer()

stemmed = [porter.stem(word) for word in no_stop_words]
stemmed
```

```python
# There are also lemmatization tools in NLTK, which help group and consolidate terms. 
# For example, "better" has the word "good" as its lemma, and "was" has "be." We will talk more about lemmatization with spaCy.
```

Using spaCy

```python
# While NLTK is a great library, another that has grown popular for its scalability and efficiency is spaCy. 
# We'll cover a few of its features here.

# First install spaCy as well as its English model. After that, you should be set to run spacy.
```

```python
!python -m spacy download en_core_web_sm
import spacy 
nlp = spacy.load("en_core_web_sm")
nlp
```

```python
#Let's load up Sleepy Hollow but this time into a spaCy doc.
```

```python
sleepy_hollow = nlp(text)
type(sleepy_hollow)
```

```python
#We can traverse the text tokens.
```

```python
[token.text for token in sleepy_hollow]
```

```python
#We can also traverse the sentences, which are packaged into Span objects.
[token.text for token in sleepy_hollow.sents]
```

```python
# There are a lot of helpful attributes with each token in spaCy. Below we iterate a handful of tokens from the Sleepy Hollow 
# document and print a few attributes we learned about previously.
for token in sleepy_hollow[50:60]: 
    print(f"Index: {token.idx}")
    print(f"Text: {token.text}")
    print(f"Is Alpha: {token.is_alpha}")
    print(f"Is Punctuation: {token.is_punct}")
    print(f"Is Stop Word: {token.is_stop}\n\n")
```

```python
# You can also implement your own tokenization procedures but we will keep the scope focused for now. 
# Let's take a look at the lemmatization of each token. Sure enough, spaCy will find the lemma of each word.
for token in sleepy_hollow: 
    if token.is_alpha:
        print(f"{token.text} -> {token.lemma_}")
```
